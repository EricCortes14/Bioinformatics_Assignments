```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

#Set the plot style
plt.style.use('seaborn-whitegrid')

#Import the file as a tsv file and add to dataframe
df = pd.read_csv("Regression-Data.tsv", sep='\t')
```

```python
#This is the plot of the Gene Profile Size vs Semantic Similarity 
%matplotlib inline

#Set plot title label, x-axis label, and y-axis label
plt.title("Gene Profile Size vs. Semantic Similarity")
plt.xlabel('Gene Profile Size')
plt.ylabel('Semantic Similarity')

#Plot the graph using the dataframe
plt.plot(df[['Gene Profile Size']], df['Semantic Similarity'], 'o', color = "blue")
```

```python
#Make linear regression model
model = LinearRegression()

#Make the model according to the specified data from the table
model.fit(df[['Gene Profile Size']], df['Semantic Similarity'])

#Obtain the r^2 value of the model
r_squared = model.score(df[['Gene Profile Size']], df['Semantic Similarity'])

#Print the r^2 result
print("The r^2 value for the Gene Profile Size vs. the Semantic Similarity model is: ", r_squared)
```

```python
#Make linear regression line
y_predicted = model.predict(df[['Gene Profile Size']])

#Plot original graph
plt.plot(df[['Gene Profile Size']], df['Semantic Similarity'], 'o', color = 'blue')

#Plot linear regression line
plt.plot(df[['Gene Profile Size']], y_predicted, color = 'red')

#Set plot title label, x-axis label, and y-axis label
plt.title("Linear Regression Line")
plt.xlabel("Gene Profile Size")
plt.ylabel("Semantic Similarity")
```

```python
#Making the residuals
residuals = df['Semantic Similarity'] - y_predicted

#Ploting the residuals on a plot
plt.plot(df[['Gene Profile Size']], residuals, 'o', color='blue')

#Set plot title label, x-axis label, and y-axis label
plt.title("Residual Plot")
plt.xlabel("Gene Profile Size")
plt.ylabel("Residuals")
```
