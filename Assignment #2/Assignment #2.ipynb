```python
# This method computes and returns the zValue that is
# needed for the z-algorithm, of an index and
# as well as string composed of a pattern and text which
# are both passed in as parameters.

def getZValue(index, patternAndText):
    
    #Variables
    zValue = 0
    
    #Loop that starts at the given index and ends at the length of patternAndText
    for x in range(index, len(patternAndText)):
        
        #If the character at x and at the current zValue of patternAndText are the same zValue is incremented
        if patternAndText[x] == patternAndText[zValue]:
            zValue += 1
            
        #If the character at x and at the current zValue are not the same the loop stops
        else:
            break
    
    #Return the zValue
    return zValue
```

```python
# This method confirms and displaysif a pattern if 
# found in a text. This method recieves two 
# parameters. First, a pattern, then a text.
def zAlgorithm(pattern, text):
    
    #Variables and Array
    patternLength = len(pattern)
    patternAndText = pattern + "$" + text
    patternAndTextLength = len(patternAndText)
    containsPattern = False
    zArray = ["x"]
    
    #Loop that finds the zValue for each character of the string patternAndText
    for x in range(1, patternAndTextLength):
        
        #Call getZValue to get zValue for x and add it to the zArray
        zValue = getZValue(x, patternAndText)
        zArray.append(zValue)
    
    #If statement that checks if length of pattern is in the zArray if it is then text contains pattern and displayed
    if patternLength in zArray:
        
        #Display for if pattern is in text
        index = zArray.index(patternLength)
        index = index - (len(pattern) + 1)
        print("Z-Algorithm:")
        print("The pattern is found in the text.")
        print("The index of the text were the pattern starts is: ", index)
        print("The Z-Array is: ", zArray, "\n")
        
    else:
        
        #Display if pattern is not in text
        print("Z-Algorithm:")
        print("The pattern is not found in the text")
        print("The Z-Array is: ", zArray, "\n")
```

```python
# This method returns the length of the longest
# string that is common in both the sets of prefixes   
# and suffixes of the original string passed in.
def getPrefixNumber(string):
    
    #Variables
    stringLength = len(string)
    result = 0
    
    #Loop that obtains prefix number
    for x in range(1, stringLength):
        
        #Obtains both prefix and suffix
        prefix = string[:x]
        suffix = string[-x:]
        
        #If statement to compare to see if prefix and suffix are equal
        if prefix == suffix:
            
            #If the length of the string is greater then the previous it gets replaced
            if len(prefix) > result:
                result = len(prefix)
    
    #Return the length of longest string
    return result
```

```python
# This method makes and returns a prefix array which contains
# the prefix number for each index of the string that is 
# passed in as a parameter.
def getPrefixArray(string):
    
    #Make a new empty array
    prefixArray = []
    
    #Loop to populate array with prefix numbers
    for x in range(1, len(string) + 1):
        
        #Method getPrefixNumber is called to obtain prefix number then added to the array
        prefixArray.append(getPrefixNumber(string[:x]))
    
    #Returns the prefix array
    return prefixArray
```

```python
# This method compares a pattern to a text, character by character
# and returns the index of the first occurance were the text and 
# pattern do not have the same character at the same index
def compareStrings(pattern, text):
    
    #Variables
    patternLength = len(pattern)
    currentIndex = 0
    
    #Loop to compare each character of the strings
    for x in range(0, patternLength):
        
        #If characters at index x are the same the varialbe currentIndex increases by one
        if pattern[x] == text[x]:
            currentIndex += 1
        
        #If characters are not the same loop is stopped
        else:
            break
    
    #The currentIndex at which the loop stopped is returned
    return currentIndex
```

```python
# This method confirms and displaysif a pattern if 
# found in a text. This method recieves two 
# parameters. First, a pattern, then a text.
def KMP(pattern, text):
    
    #Obtains the prefixArray by calling getPrefixArray
    prefixArray = getPrefixArray(pattern)
    
    #Variables
    containsPattern = False
    textIndex = 0
    finalTextLength = len(text) - len(pattern)
    patternArray = []
    indexArray = []
    
    #Loop that 
    while textIndex <= finalTextLength:
        
        #Obtains index of the first character where the pattern and text are not the same 
        compareStringsNumber = compareStrings(pattern, text[textIndex:])
        
        #If the compareStringsNumber is not eaual to the length of the patter then the index of the text is adjusted
        if compareStringsNumber != len(pattern):
            
            #Find how many indices the index of the text should be moved
            adjustmentNumber = prefixArray[compareStringsNumber]
            
            #If the adjustmentNumber is 0 then it moves one spot
            if adjustmentNumber == 0:
                textIndex += 1
            
            #If the adjustmentNumber is not 0 the adjustmentNumber is added to the previous index of the text
            else:
                textIndex = textIndex + adjustmentNumber
        
        #Else the length of the compareStringsNumber is the same as the legnth of the pattern
        #which means that the pattern was found in the text
        else:
            
            #Set containsPattern to true then stops the loop
            containsPattern = True
            break
            
    if containsPattern == True:
        print("KMP Algorithm:")
        print("The pattern is found in the text.")
        print("The index of the text were the pattern starts is: ", textIndex)
    else:
        print("KMP Algorithm:")
        print("The pattern is not found in the text")
    
    #Display prefix table
    for x in range(0, len(pattern)):
        patternArray.append(pattern[x])
        indexArray.append(x)
    print("Prefix Table:")
    print("i:", indexArray)
    print("Text[i]:", patternArray)
    print("Prefix[i]:", prefixArray)
```
